## First Home Bank Analysis

```python
fhb = loans.Bank("First Home Bank")
```

## Q1: what is the average interest rate for the bank?

```python
#Q1
summ = 0
total = 0

for l in fhb.loans:
    num = float(str(l).split(" ")[1].split("%")[0])
    if num != -1:
        summ += num
        total += 1

summ / total
```

## Q2: how many applicants are there per loan, on average?

```python
#Q2
summ = 0
total = 0

for l in fhb.loans:
    app_num = int(str(l).split(" ")[5])
    summ += app_num
    total += 1
    
summ / total
```

## Q3: what is the distribution of ages?

```python
#Q3
ages_d = {}

for loan in fhb.loans:
    for app in loan.applicants:
        if app.age in ages_d:
            ages_d[app.age] += 1
        else:
            ages_d[app.age] = 1
                     
ages_d        
```

## Q4: how many interest rate values are missing?

```python
#Q4
t = search.BST()
for l in fhb.loans:
    t.add(l.interest_rate, l) 

len(t[-1])
```

## Q5: how tall is the tree?

```python
#Q5
def max_depth(node):
    if node is None or (node.left is None and node.right is None):
        return 1
    return 1 + max(max_depth(node.left), max_depth(node.right))

max_depth(t.root)
```

## University of Wisconsin Credit Union Analysis
## Q6: how long does it take to add the loans to the tree?

```python
#Q6
uwcu = loans.Bank("University of Wisconsin Credit Union")

times_df = pd.DataFrame(dtype = float)

t = search.BST()
t0 = time.time()
counter = 0

for l in uwcu.loans:
    t.add(l.interest_rate, l)
    counter += 1 
    t1 = time.time()
    running_time = (t1 - t0) * 1000
    times_df.at[counter, "Time"] = running_time
    
df = times_df.plot.line(legend = False)
df.set_xlabel("BST Adds")
df.set_ylabel("Total Elapsed Milliseconds")  
None
```

## Q8: what is the relationship between property value and loan amount?

```python
#Q8
loans_df = pd.DataFrame(dtype = int)
for l in uwcu.loans:
    value = int(l.property_value)
    if value > 1000000 or value < 0:
        continue
    amt = int(l.loan_amount)
    loans_df.at[l, "value"] = value
    loans_df.at[l, "amount"] = amt
    
ax = loans_df.plot.scatter(x = "value", y = "amount", alpha = 0.01, color = "black")
ax
```

## Q9: what is the distribution of race for UWCU loan applicants?

```python
#Q9
races = {}
races['White'] = 0
races['Unknown'] = 0
races['2+'] = 0

for l in uwcu.loans:
    apps = l.applicants
    for a in apps:
        race = list(a.race)
        if len(race) == 0:
            races["Unknown"] += 1
            continue
        if len(race) == 1:
            if not race[0] in races:
                races[race[0]] = 1
            else:
                races[race[0]] += 1
        else:
            races["2+"] += 1
           
ax = pd.DataFrame.from_dict(races, orient = 'index')
ax = ax.plot.bar(legend = False)
ax.set_ylabel("Number of Applicants")
ax.set_xlabel("Race")
```
